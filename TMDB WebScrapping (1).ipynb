```python
url_main = 'https://www.themoviedb.org/movie'
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'
}
```

```python
url_list = [url_main]
for i in range(1, 51):  
    temp_url = f'{url_main}?page={i}'
    url_list.append(temp_url)
```

```python
url_list
```

```python
final_movie_list = []
for url in url_list:
    print(url)
    print('=======================================================')
    source_data = requests.get(url, headers=header).text
    soup_data = BeautifulSoup(source_data, 'lxml')

    movies = soup_data.find_all('div', class_='card style_1')
    movies_list = []
    for movie_elem in movies:
        exp_str = str(movie_elem.find('a'))[25:40]  
        
        base = "https://www.themoviedb.org/movie/"
        digits = re.findall("\d", exp_str)
        ext = "".join(digits)
        movie_link = base + ext
        #movies_list.append(movie_link)
        
        
        # Fetch data for each movie link
        source_movie_data = requests.get(movie_link, headers=header).text
        soup_movie_data = BeautifulSoup(source_movie_data, 'lxml')

        # Extracting movie details
        try:
            movie_name = soup_movie_data.find('h2').find('a').text
        except:
            movie_name = None
        
        try:
            release_date_1 = soup_movie_data.find('span', {'class': 'release'}).text
            release_date = re.findall(r'\d{2}.\d{2}.\d{4}', release_date_1)[0]
        except:
            release_date = None
        
        try:
            rating = float(soup_movie_data.find('div', {'class': 'user_score_chart'})["data-percent"])
        except:
            rating = None

        # Extract director information
        
        try:
        
            cast_members = soup_movie_data.find('ol', {'class': 'people no_image'}).find_all('li')
            director = None
            for person in cast_members:
                person_str = str(person)
                if 'Director' in person_str:
                    director = person.find('a').text
                    break
        
        except:
            director = None
        
        
        try:
            runtime = soup_movie_data.find('span', class_='runtime').text.replace('\n', '').strip()
        except:
            runtime = None
        try:
            genre = soup_movie_data.find('span', class_='genres').text.replace('\n', '').replace('\xa0', '')
        except:
            genre = None
        
        # Append movie details to the movies_list
        movies_list.append({
            'Movie Name': movie_name,
            'Release Date': release_date,
            'User Rating': rating,
            'Director': director,
            'Runtime': runtime,
            'Genres': genre
        })
        print(movies_list[-1])
        
    final_movie_list.extend(movies_list)
```

```python
res = pd.DataFrame(final_movie_list)
res.to_excel('new.xlsx')
```
